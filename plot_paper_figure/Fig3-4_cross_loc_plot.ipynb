```python
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### FUS-miRNA ALEX

```python
folder_FUS_miR21 = "/Volumes/lsa-nwalter/Guoming_Gao_turbo/Walterlab_server/PROCESSED_DATA/RNA-diffusion-in-FUS/RNAinFUS_PaperFigures/Fig4_regulation/FUS488_miR21_PAINT/example_condensate_FUS-miR21/"
os.chdir(folder_FUS_miR21)
df_FL = pd.read_csv("red.csv")
df_FUS = pd.read_csv("cyan.csv")

distance_um = df_FL['Distance_(microns)'].to_numpy(float)
red = df_FL['Gray_Value'].to_numpy(float)
cyan = df_FUS['Gray_Value'].to_numpy(float)

fontsize = 38

plt.figure(figsize=(12, 3))
plt.plot(distance_um, cyan, color="cyan", lw=5, label="FUS")
plt.plot(distance_um, red, color="red", lw=5, label="miRNA")
plt.xlim(distance_um[0], distance_um[-1])
plt.legend(frameon=False, fontsize=fontsize, loc=7, bbox_to_anchor=(1.43, 0.5))
plt.xticks([0, 1, 2, 3])
plt.ylim(0, 25)
plt.yticks([0, 10, 20])
plt.text(3.8, -4.5, "Distance, $\mu$m", fontsize=fontsize)
plt.xlabel("")
plt.ylabel("#Loc.\n/pxl.", fontsize=fontsize)
plt.gca().spines[:].set_linewidth(2)
plt.gca().tick_params(
    axis="both",
    which="major",
    labelsize=fontsize,
    direction="in",
    bottom=True,
    left=True,
    length=5,
    width=2,
)
plt.savefig("cross_section_redcyan.png", format="png", dpi=600, bbox_inches="tight")
```

### FUS-FL ALEX stepsize

```python
folder_FUS_FL = "/Volumes/lsa-nwalter/Guoming_Gao_turbo/Walterlab_server/PROCESSED_DATA/RNA-diffusion-in-FUS/RNAinFUS_PaperFigures/Fig3_coralled by nano domains/FUS488_FL_PAINT/example_condensates_FUS-FL"
os.chdir(folder_FUS_FL)
df_FL = pd.read_csv("FOV-5-78-stepsize-FL.csv")
df_FUS = pd.read_csv("FOV-5-78-stepsize-FUS.csv")

distance_um = df_FL["Distance_(microns)"].to_numpy(float)
magenta = df_FL["Gray_Value"].to_numpy(float)*1000
yellow = df_FUS["Gray_Value"].to_numpy(float)*1000

fontsize = 38

plt.figure(figsize=(12, 3))
plt.plot(distance_um, yellow, color="orange", lw=5, label="FUS")
plt.plot(distance_um, magenta, color="magenta", lw=5, label="mRNA")
plt.legend(frameon=False, fontsize=fontsize, loc=7, bbox_to_anchor=(1.43, 0.5))
plt.xlim(0, 3.5)
# plt.xticks([0, 1, 2, 3])
plt.ylim(135, 215)
plt.text(3.51, 119, "Distance, $\mu$m", fontsize=fontsize)
plt.xlabel("")
plt.ylabel("Step, nm", fontsize=fontsize)
plt.gca().spines[:].set_linewidth(2)
plt.gca().tick_params(
    axis="both",
    which="major",
    labelsize=fontsize,
    direction="in",
    bottom=True,
    left=True,
    length=5,
    width=2,
)
plt.savefig("cross_section_step_yellowFUSmagentaFL.png", format="png", dpi=600, bbox_inches="tight")
```
