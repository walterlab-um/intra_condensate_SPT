### Assemble phase diagram

```python
df_plot = df[df["salt_type"] == "KCl"]
```

```python
plt.figure(figsize=(3, 3), dpi=600)
ax = sns.scatterplot(
    data=df_plot,
    x="FUS_uM",
    y="salt_M",
    hue="LLPS",
    palette=["#a7cced", "#304d6d"],
    hue_order=["n", "y"],
    s=200,
    legend=False,
)
plt.gca().spines[:].set_linewidth(2)
plt.xticks([0, 1, 2, 3, 4, 5])
# plt.yticks([0.1, 0.2])
plt.gca().tick_params(
    axis="both",
    which="major",
    labelsize=17,
    direction="in",
    bottom=True,
    left=True,
    length=5,
    width=2,
)
plt.xlabel(
    r"FUS, $\mu$M",
    weight="bold",
    fontsize=17,
)
plt.ylabel(
    "KCl, M",
    weight="bold",
    fontsize=17,
)
plt.xlim(0, 5.5)
plt.ylim(0, 1.1)
# sns.move_legend(ax, 0, title=None, frameon=False, ncols=2)
plt.savefig("Fig1_phase_diagram.png", format='png', bbox_inches="tight")
plt.show()
```
